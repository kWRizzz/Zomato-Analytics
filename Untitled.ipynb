```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
df_final=pd.merge(df,df_country,on="Country Code",how="left")
df_final.head(2)
```

```python
df_final.info()
```

```python
df_final.Country.value_counts()
# ab tumne dekha ki sabse jada india me zomato operate ho rha hai toh ab tumhe iska chart bana hai .index karpo to list of string dedega
```

```python
countries=df_final.Country.value_counts().index
```

```python
# ab iske har country ke value bhi daal rha hu 
countries_values=df_final.Country.value_counts().values
countries_values
```

```python
# top 3 countries ka nikal rhe hai iss liye [:3] lagay hai bhai
plt.pie(countries_values[:3],labels=countries[:3], autopct="%1.2f%%")
```

 # observations: now the record says that majority are  the india then usa then uk 

```python
df_final.columns
```

```python
ratings=df_final.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:"Rating count"})
```

# Observation 
* rating between [0,2.5) POOR
* rating between [2.5,3.5) AVERAGE
* rating between [3.5,4.0) GOOD
* rating between [4.0,4.5) VERRY GOOD
* rating between [4.5,5.0) EXCELLENT

```python
ratings
```

```python
plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings)
```
